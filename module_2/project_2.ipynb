Видно, что в ряде строковых столбцов присутствуют значения `NaN`. Это необходимо исправить. 
Создадим функцию, которая заменит в указанном строковом столбце все пустые значение, а также значения `NaN` в любом виде на `None` для удобства дальнейшей обработки

```python
def fix_str_column(column):
    # создадим дополнительную функцию для упрощения понимания кода
    def is_bad_str(x):
        # не забываем преобразовать значение в строку
        s = str(x)
        # возвращаем True, строка пустая или содержит NaN в любом виде
        return (s.lower() == 'nan') or (s == '')

    df[column] = df[column].apply(lambda x: None if pd.isnull(
        x) else None if is_bad_str(x) else x)
```

Теперь очистим от мусора значения во всех строковых столбцах

```python
# Объявим вспомогательную функцию, которая будет вызывать заданный коллбек для каждого столбца заданного типа
def for_columns_by_type(callback, col_type='object'):
    col_types = df.dtypes
    for column in col_types.index:            # пробегаем по всем столбцам
        if col_types[column] == col_type:     # и если столбец соответствует заданному типу
            # вызываем заданный коллбек
            callback(column)
```

```python
for_columns_by_type(fix_str_column)
df.head(10)
```

Объявим функцию для проверки столбца на наличие недействительных значений

```python
# Параметр 'column' - в каком столбце искать недействительные значения
# Параметр 'counter_column' - по какому столбцу считать недействительные значения
# Нельзя указывать одинаковые, т.к. по неведомой для меня причине результат будет всегда 0
def are_values_finite(column, counter_column='score'):
    if column == counter_column:
        print('В столбце \'', column,
              '\' невозможно посчитать кол-во недействительных значений', sep='')
        return

    # кочество чисел, исключая NaN и Infinity
    f_cnt = df[counter_column][pd.isna(df[column])].count()
    if f_cnt == 0:
        print('В столбце \'', column, '\' нет недействительных значений', sep='')
    else:
        print('В столбце \'', column, '\' ', f_cnt,
              ' недействительных значений', sep='')
```

## Проверка значений номинативных столбцов

Объявим функцию, которая будет сигнализировать о наличии статистически значимой зависимости между значениями указанного строкового столбца и столбца `score`

```python
def get_stat_dif(column):
    # для начала вычленим все комбинации значений внутри столбца
    cols = df[column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    # вычислим поправку Бонферони для теста Стьюдента
    bonf = 0.05 / len(combinations_all)
    # прогоним тест Стьюдента для всех найденных комбинаций
    for comb in combinations_all:
        if ttest_ind(df.loc[df.loc[:, column] == comb[0], 'score'],
                     df.loc[df.loc[:, column] == comb[1], 'score']).pvalue <= bonf:
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
for_columns_by_type(get_stat_dif)
```

ОК, есть еще 4 столбца, значения которых имеют статистически значимое влияние на значение `score`. Тем не менее, считаю необходимым дополнительно исследовать столбцы, имещие бинарные значения (например, `yes` и `no`). Преобразуем такие стобцы в числовые, где одному из значений будет приравнена `1`, а другому `0`. Для удобства создадим новый датафрейм.

```python
# создаем датафрейм на основе столбца score
df2 = pd.DataFrame(df.score)

# объявим метод, который будет добавлять в новый датафрейм
# новый столбец на основании бинарного столбца из исходного дата-фрейма


def translate_to_num(column, value_of_one):
    # дополнительно проверим, что столбец действительно бинарный
    # т.е. имеет всего 2 возможных значения
    vci = df[column].value_counts().index
    if len(vci) != 2:
        print('столбец \'', column, '\' не является бинарным', sep='')
        return

    df2[column] = df[column].apply(
        lambda x: 1 if str(x) == value_of_one else 0)


translate_to_num('sex', 'M')
translate_to_num('address', 'U')
translate_to_num('fam_size', 'GT3')
translate_to_num('p_status', 'T')

for col in ['school_sup', 'fam_sup', 'paid', 'activities', 'nursery', 'internet', 'romantic']:
    translate_to_num(col, 'yes')
```

Теперь рассмотрим корреляцию полученных значений с искомыми (столбцом `score`)

```python
df2.corrwith(df2.score).sort_values()
```

Результат не противоречит тесту Стьюдента, однако, столбец `sex` тест Стьюдента не прошел, хотя имеет значение корреляции, близкое к столбцу `address`. Считаю, что столбец `sex` должен быть добавлен в данные для анализа, т.к. возможна более глубокая взаимосвязь.
